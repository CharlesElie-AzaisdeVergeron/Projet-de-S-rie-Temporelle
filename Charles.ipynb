```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression,Ridge,Lasso
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("hour.csv")

# Remove rows with missing values
data = data.dropna()

# Split the data into training and test sets
n = int(0.8 * data.shape[0])
train = list(range(n))
test = list(range(n, data.shape[0]))

data_train = data.iloc[train]
data_test = data.iloc[test]


y_train = data_train['cnt'].values
X_train = data_train.drop(columns=['dteday','cnt']).values
X_test = data_test.drop(columns=['dteday','cnt']).values
y_test = data_test['cnt'].values
```

```python
LR = LinearRegression()
LR.fit(X_train, y_train)
y_pred = LR.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (Linear Regression): {mse}")
```
